```python
url="https://www.91wheels.com/compare-car/hyundai-creta-vs-kia-seltos-vs-mahindra-scorpio-classic"
```

```python
page=requests.get(url) 
page
```

#### "requests.get" is used to send an HTTP GET request to a specified URL, allowing you to retrieve data and content from that web address.if we get response as 200 then web site allows us to retrieve data if we get responses other than 200 i.e 404,500 that means the respective website does not give permission to retrieve the data

```python
content=bs(page.text)
content.prettify
```

#### bs represents a BeautifulSoup object, typically obtained after parsing an HTML or XML document. bs.text is a method used to extract the text content from the parsed document. It returns all the textual content within the HTML or XML, stripping away any HTML tags and leaving you with just the plain text.

```python
Title=content.title
Title
```

```python
Title.string
```

Here with the help of title attribute i access the text inside the title tag 

```python
cars=content.find_all("div",class_="font-12 lineHtSmall d-flex-in pos-r")
Cars=[]
for i in cars:
    Cars.append(i.get_text())
for i in Cars:
    print(i)
```

here i retrieve the names of the cars that i wanted to compare,in the list called Cars.


```python
carimgs=content.find_all("img",class_="")
img=[]
for i in carimgs:
    img.append(i["src"])
del img[3:128]

for l in img:
    print (l)
```

here i retrieve the links of images of the cars in the list img.

```python
img
```

```python
a={"Car_imgs":img,"Cars":Cars}
Car_comparison=pd.DataFrame(a)
Car_comparison
```

"find_all" is a method provided by BeautifulSoup to locate and extract all HTML elements on a web page that match a specific HTML tag or set of criteria. using find all method first i retrive all the data inside the table body and after that i differentiate and the rows inside the table body

```python
table=content.find_all("tbody",class_="")
table
```

```python
rows=content.find_all('tr',class_="")
rows
```

```python
column_info=[]
for i in rows:
    cols=i.find_all('td')
    Car_info=[c.text for c in cols]
    column_info.append(Car_info)
    print(column_info)
```

here i have extracted all the data present in the table rows and defined them in variable called column_info.

```python
column_info
```

```python
column_info.pop(1)
```

 The data i got by extracting the rows had empty list so that i decided to remove that list to avoid any dimensionality problem.

```python
Features=[]
Car1=[]
Car2=[]
Car3=[]
for i in column_info :
    Features.append(i[0])
    Car1.append(i[1])
    Car2.append(i[2])
    Car3.append(i[3])
```

after extracting all the data i tried to append all the information related to each car into respected variable

```python
Data={"Features":Features,"Car1":Car1,"Car2":Car2,"Car3":Car3}
```

storing data in dictionary format in variable called data.

```python
Cars_Comparison=pd.DataFrame(Data)
Cars_Comparison
```

 Converting the data stored in variable called data into tabular format called dataframe using pandas library

```python
Cars_Comparison.to_csv("Car_comparison_data",index=False )
```
